```python
import nibabel as nib
import nibabel.processing as nibproc
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
dir_path = os.getcwd() + "/backup/my_pix2pix/Task1/train/"
list_files = sorted(os.listdir(dir_path))
# print(list_files)

train = []
for file in list_files[:3]:
    cur = {}
    src = dir_path+file+"/mr.nii.gz"
    target = dir_path+file+"/ct.nii.gz"

    cur['id'] = file
    
    nii_img = nib.load(src)
    nii_img = nibproc.resample_from_to(nii_img, [(256, 256, 160), nii_img.affine])
    cur['src'] = np.array(nii_img.get_fdata()) 

    nii_img = nib.load(target)
    nii_img = nibproc.resample_from_to(nii_img, [(256, 256, 160), nii_img.affine])
    cur['target'] = np.array(nii_img.get_fdata())

    train.append(cur)
```

```python
img = train[0]['src'][:, : , 100:101]
img = img.transpose(2,0,1)
plt.imshow(img, cmap='gray')  # Display the middle slice of the 3D volume
plt.colorbar()
plt.title("MRI Scan")
plt.show()
```
