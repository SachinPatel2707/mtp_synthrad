```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Read data from CSV file into a DataFrame
file_path = "Models/v1.5.3/evaluation/losses.csv"  # Replace with the actual path to your CSV file
df = pd.read_csv(file_path)

# Step 3: Create subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 8))

# Plot for x column
axs[0].plot(df.index, df['D_loss'], label='D_loss', color='blue')
axs[0].set_title('Discriminator_loss')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('D_loss')
# axs[0].legend()

# Plot for y column
axs[1].plot(df.index, df['G_loss'], label='G_loss', color='orange')
axs[1].set_title('Generator_loss')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('G_loss')
# axs[1].legend()

# Adjust layout to prevent clipping of titles
plt.tight_layout()

# Show the plots
plt.show()
```
